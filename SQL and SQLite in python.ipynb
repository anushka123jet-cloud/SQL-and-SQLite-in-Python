```python
import sqlite3
```

```python
#Connect to an SQLite database
connection = sqlite3.connect('example.ds')
connection
```

```python
cursor = connection.cursor()
```

```python
#Create a table
cursor.execute('''
Create Table if not Exists employees(
       id Integer Primary Key,
       name Text Not Null,
       age Integer,
       department text
       )
''')
```

```python
#Commit changes
connection.commit()
```

```python
cursor.execute('''
Select * from employees
''')
```

```python
#Insert the data in sqlite table
cursor.execute('''
Insert Into employees(name,age,department)
            values('Krish',32,'Data Scientist')
            ''')

cursor.execute('''
Insert Into employees(name,age,department)
            values('Anushka',23,'Computer Science')
            ''')

cursor.execute('''
Insert Into employees(name,age,department)
            values('Tripty',24,'Software Developer')
            ''')
#commit changes
connection.commit()
```

```python
#Query for data from the table
cursor.execute('Select * from employees')
rows = cursor.fetchall()
```

```python
#Print the queried data
for row in rows:
    print(row)
```

```python
#Update the data in the table
cursor.execute('''
UPDATE employees
Set age == 34
where name == "Krish"
''')
connection.commit()
cursor.execute('Select * from employees')
for row in rows:
    print(row)
```
